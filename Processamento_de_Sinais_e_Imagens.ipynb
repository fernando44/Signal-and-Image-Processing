```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('/content/exe.jpg')

edges = cv2.Canny(image, 160, 160)

plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Imagem Original')
plt.show()
plt.subplot(121), plt.imshow(edges, cmap='gray'), plt.title('Detecção de bordas')
plt.show()
```
